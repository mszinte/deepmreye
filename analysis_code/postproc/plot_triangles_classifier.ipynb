```python
def adapt_evaluation(participant_evaluation):
    pred_y = participant_evaluation["pred_y"]
    pred_y_median = np.nanmedian(pred_y, axis=1)
    pred_uncertainty = abs(participant_evaluation["euc_pred"])
    pred_uncertainty_median = np.nanmedian(pred_uncertainty, axis=1)
    df_pred_median = pd.DataFrame(
        np.concatenate(
            (pred_y_median, pred_uncertainty_median[..., np.newaxis]), axis=1),
        columns=["X", "Y", "Uncertainty"],
    )
    # With subTR
    subtr_values = np.concatenate((pred_y, pred_uncertainty[..., np.newaxis]),
                                  axis=2)
    index = pd.MultiIndex.from_product(
        [range(subtr_values.shape[0]),
         range(subtr_values.shape[1])],
        names=["TR", "subTR"])
    df_pred_subtr = pd.DataFrame(subtr_values.reshape(-1,
                                                      subtr_values.shape[-1]),
                                 index=index,
                                 columns=["X", "Y", "pred_error"])

    return df_pred_median, df_pred_subtr
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

slice_length = 9
nTR_run = 313

task_TRs = {
    "task_1": (5, 77),
    "task_2": (82, 154),
    "task_3": (159, 231),
    "task_4": (236, 308),
}

# Define your subject DataFrames (
subject_dfs = {
    1: df_pred_median_1_calib,
    2: df_pred_median_2_calib,
    3: df_pred_median_3_calib,
    4: df_pred_median_4_calib,
    6: df_pred_median_6_calib,
    7: df_pred_median_7_calib,
    9: df_pred_median_9_calib,
    10: df_pred_median_10_calib,
    11: df_pred_median_11_calib,
    13: df_pred_median_13_calib,
    14: df_pred_median_14_calib,
    15: df_pred_median_15_calib, 
    16: df_pred_median_16_calib,
    17: df_pred_median_17_calib
}

# Labeling pattern (TR index, label)
pattern = [
    (5, 'up'), (41, 'up'), (82, 'up'), (118, 'up'), (159, 'up'), (195, 'up'), (236, 'up'), (272, 'up'),
    (14, 'right'), (50, 'right'), (91, 'right'), (127, 'right'), (168, 'right'), (204, 'right'), (245, 'right'), (281, 'right'),
    (23, 'down'), (59, 'down'), (100, 'down'), (136, 'down'), (177, 'down'), (213, 'down'), (254, 'down'), (290, 'down'),
    (32, 'left'), (68, 'left'), (109, 'left'), (145, 'left'), (186, 'left'), (222, 'left'), (263, 'left'), (299, 'left')
]

# Shift TRs by +1 (to form full triangles)
indices_and_labels = []
for i in range(3):  # 3 runs
    #indices_and_labels.extend([(tr + i * nTR_run + 2, label) for tr, label in pattern])  
    indices_and_labels.extend([(tr + i * nTR_run, label) for tr, label in pattern])

# Create slices
all_data = []
tr_indices = []

for subject_id, df in subject_dfs.items():
    df = df.copy()
    df['subject'] = subject_id

    for start_index, label in indices_and_labels:
        end_index = start_index + slice_length
        if end_index <= len(df):
            x_slice = df['X'].iloc[start_index:end_index].values
            y_slice = df['Y'].iloc[start_index:end_index].values

            if len(x_slice) == slice_length and len(y_slice) == slice_length:
                all_data.append([subject_id, x_slice, y_slice, label])
                tr_indices.append(start_index)

# Convert to DataFrame
columns = ['subject'] + [f'x{i+1}' for i in range(slice_length)] + [f'y{i+1}' for i in range(slice_length)] + ['label']
formatted_data = []

for entry in all_data:
    subject_id, x_slice, y_slice, label = entry
    formatted_data.append([subject_id] + list(x_slice) + list(y_slice) + [label])

df_final = pd.DataFrame(formatted_data, columns=columns)

# Encode labels
label_encoder = LabelEncoder()
df_final['label_encoded'] = label_encoder.fit_transform(df_final['label'])

label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Label Mapping:", label_mapping)

# Task assignment function
def get_task_label(tr):
    tr_in_run = tr % nTR_run
    for task, (start, end) in task_TRs.items():
        if start <= tr_in_run < end:
            return task
    return "ITI"

df_final["task"] = [get_task_label(tr) for tr in tr_indices]

print(df_final.head())
```

```python
import random

# Randomize within each sample
def shuffle_sequence(row):
    coords = list(zip(
        [row[f'x{i}'] for i in range(1, slice_length + 1)],
        [row[f'y{i}'] for i in range(1, slice_length + 1)]
    ))
    random.shuffle(coords)  # shuffle in place
    xs, ys = zip(*coords)   # unzip back
    for i in range(slice_length):
        row[f'x{i+1}'] = xs[i]
        row[f'y{i+1}'] = ys[i]
    return row

# Apply to all rows
df_shuffled = df_final.copy().apply(shuffle_sequence, axis=1)
```

```python
from matplotlib.lines import Line2D
from matplotlib.colors import to_rgb, to_hex

def generate_discrete_colormap(start_hex, end_hex, n_steps):
    start_rgb = np.array(to_rgb(start_hex))
    end_rgb = np.array(to_rgb(end_hex))
    return [to_hex(start_rgb + (end_rgb - start_rgb) * i / (n_steps - 1)) for i in range(n_steps)]

def plot_colored_sequence_discrete(ax, x_vals, y_vals, title):
    num_points = len(x_vals)
    colors = generate_discrete_colormap("#A8C9A9", "#3D6B46", num_points)

    # Create segments for the lines between points
    points = np.array([x_vals, y_vals]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)

    # Line colors based on start point of each segment
    lc = LineCollection(segments, colors=colors[:-1], linewidths=2)
    ax.add_collection(lc)

    # Legend for steps
    legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[i],
                              markeredgecolor='black', label=f'Step {i+1}', markersize=8)
                       for i in range(num_points)]
    ax.legend(handles=legend_elements, title="Steps", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)

    ax.set_title(title)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.set_xlim(min(x_vals) - 1, max(x_vals) + 1)
    ax.set_ylim(min(y_vals) - 1, max(y_vals) + 1)
```
